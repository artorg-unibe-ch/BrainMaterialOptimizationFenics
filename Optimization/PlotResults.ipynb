```python
#########################################################################
### Plot the evolution of the parameter during optimization procedure ###
#########################################################################


import pandas as pd                       # Allows to use data in tables
import matplotlib.pyplot as plt           # Allows to do plots


NumberElementsTested = [1,2,3,5,10,20]
InitialGuess = ['BuddayParameters', 'MarcParameters']

for NumberElements in NumberElementsTested:

    dfB = pd.read_csv('OptimizationResults/' + InitialGuess[0] + '/' + str(NumberElements) + 'Elements.txt', sep=' ', decimal='.')
    dfM = pd.read_csv('OptimizationResults/' + InitialGuess[1] + '/' + str(NumberElements) + 'Elements.txt', sep=' ', decimal='.')

    plt.rc('figure', figsize=[36,7])
    fig = plt.figure()
    fig.suptitle('Values Evolution %i Elements by side'%(NumberElements))
    ax = fig.add_subplot(1, 4, 1)
    ax.cla()
    ax.plot(dfB.Nu,  color = 'k', linestyle = '--', marker = 'o', markersize = 8, fillstyle='none', label='Budday')
    ax.plot(dfM.Nu,  color = 'gray', linestyle = '--', marker = 'o', markersize = 8, fillstyle='none', label='Marc')
    ax.set_xlabel('Iteration number(-)')
    ax.set_ylabel('Nu (-)')

    ax = fig.add_subplot(1, 4, 2)
    ax.cla()
    ax.plot(dfB.Mu,  color = 'r', linestyle = '--', marker = 'o', markersize = 8, fillstyle='none', label='Budday')
    ax.plot(dfM.Mu,  color = 'm', linestyle = '--', marker = 'o', markersize = 8, fillstyle='none', label='Marc')
    ax.set_xlabel('Iteration number(-)')
    ax.set_ylabel('Mu (kPa)')

    ax = fig.add_subplot(1, 4, 3)
    ax.cla()
    ax.plot(dfB.Alpha,  color = 'b', linestyle = '--', marker = 'o', markersize = 8, fillstyle='none', label='Budday')
    ax.plot(dfM.Alpha,  color = 'c', linestyle = '--', marker = 'o', markersize = 8, fillstyle='none', label='Marc')
    ax.set_xlabel('Iteration number(-)')
    ax.set_ylabel('Alpha (-)')

    ax = fig.add_subplot(1, 4, 4)
    ax.cla()
    ax.plot(dfB.TotalCost,  color = 'g', linestyle = '--', marker = 'o', markersize = 8, fillstyle='none', label='Budday')
    ax.plot(dfM.TotalCost,  color = 'y', linestyle = '--', marker = 'o', markersize = 8, fillstyle='none', label='Marc')
    ax.set_xlabel('Iteration number(-)')
    ax.set_ylabel('Cost (-)')
    
    ax.legend(loc='upper left', frameon=True, framealpha=1)
```

```python
#########################################################################
########### Plot the final values after optimization procedure ##########
#########################################################################


import pandas as pd                       # Allows to use data in tables
import matplotlib.pyplot as plt           # Allows to do plots


NumberElementsTested = [1,2,3,5,10,20]

Nus    = []
Mus    = []
Alphas = []
Costs  = []

for NumberElements in NumberElementsTested:

    dfB = pd.read_csv('OptimizationResults/' + InitialGuess[0] + '/' + str(NumberElements) + 'Elements.txt', sep=' ', decimal='.')
    dfM = pd.read_csv('OptimizationResults/' + InitialGuess[1] + '/' + str(NumberElements) + 'Elements.txt', sep=' ', decimal='.')
    
    FinalValuesB = dfB.iloc[-1]
    FinalValuesM = dfM.iloc[-1]
    
    FinalCostsB = dfB.iloc[-2]
    FinalCostsM = dfM.iloc[-2]

    Nus.append(FinalValuesB.Nu)
    Mus.append(FinalValuesB.Mu)
    Alphas.append(FinalValuesB.Alpha)
    Costs.append(FinalCostsB.TotalCost)
    
    Nus.append(FinalValuesM.Nu)
    Mus.append(FinalValuesM.Mu)
    Alphas.append(FinalValuesM.Alpha)
    Costs.append(FinalCostsM.TotalCost)
    
plt.rc('figure', figsize=[36,7])
fig = plt.figure('Final Values')
fig.suptitle('Final Values')
ax = fig.add_subplot(1, 4, 1)
ax.cla()
ax.plot(NumberElementsTested, Nus[::2],  color = 'k', linestyle = 'none', marker = 'o', markersize = 8, fillstyle='none', label='Budday')
ax.plot(NumberElementsTested, Nus[1::2],  color = 'gray', linestyle = 'none', marker = 'x', markersize = 8, fillstyle='none', label='Marc')
ax.set_xlabel('Number of elements by side (-)')
ax.set_ylabel('Nu (-)')
ax.set_xlim(0,21)
ax.set_ylim(0,0.55)
plt.xticks(NumberElementsTested)
ax.legend(loc='upper left', frameon=True, framealpha=1)

ax = fig.add_subplot(1, 4, 2)
ax.cla()
ax.plot(NumberElementsTested, Mus[::2],  color = 'r', linestyle = 'none', marker = 'o', markersize = 8, fillstyle='none', label='Budday')
ax.plot(NumberElementsTested, Mus[1::2],  color = 'm', linestyle = 'none', marker = 'x', markersize = 8, fillstyle='none', label='Marc')
ax.set_xlabel('Number of elements by side (-)')
ax.set_ylabel('Mu (kPa)')
ax.set_xlim(0,21)
ax.set_ylim(0,1)
plt.xticks(NumberElementsTested)
ax.legend(loc='upper left', frameon=True, framealpha=1)

ax = fig.add_subplot(1, 4, 3)
ax.cla()
ax.plot(NumberElementsTested, Alphas[::2],  color = 'b', linestyle = 'none', marker = 'o', markersize = 8, fillstyle='none', label='Budday')
ax.plot(NumberElementsTested, Alphas[1::2],  color = 'c', linestyle = 'none', marker = 'x', markersize = 8, fillstyle='none', label='Marc')
ax.set_xlabel('Number of elements by side (-)')
ax.set_ylabel('Alpha (-)')
ax.set_xlim(0,21)
ax.set_ylim(-50,-20)
plt.xticks(NumberElementsTested)
ax.legend(loc='upper left', frameon=True, framealpha=1)

ax = fig.add_subplot(1, 4, 4)
ax.cla()
ax.plot(NumberElementsTested, Costs[::2],  color = 'g', linestyle = 'none', marker = 'o', markersize = 8, fillstyle='none', label='Budday')
ax.plot(NumberElementsTested, Costs[1::2],  color = 'y', linestyle = 'none', marker = 'x', markersize = 8, fillstyle='none', label='Marc')
ax.set_xlabel('Number of elements by side (-)')
ax.set_ylabel('Cost (-)')
ax.set_xlim(0,21)
ax.set_ylim(0,0.5)
plt.xticks(NumberElementsTested)
ax.legend(loc='upper left', frameon=True, framealpha=1)

plt.show()
```

```python
import pandas as pd                       # Allows to use data in tables
import matplotlib.pyplot as plt           # Allows to do plots
import os

LoadCases = ['Compression', 'Tension', 'SimpleShear']
#LoadCase = LoadCases[0]

NumberElementsTested = [1,2,3,5,10,20]

InitialGuess = ['BuddayParameters', 'MarcParameters']
InitialGuess = InitialGuess[1]

plt.rc('figure', figsize=[34,7])
fig = plt.figure()
fig.suptitle('Simulation Results Comparison')

for LoadCase in LoadCases:
    
    df   = pd.read_csv('ExperimentalData/CR_' + LoadCase + '_ExpDat.csv', sep=';', decimal=',', header=None)
    df1  = pd.read_csv('ResultsOgden/' +InitialGuess +'/' + str(NumberElementsTested[0]) + 'Elements_' + LoadCase, sep=',', decimal='.', usecols=[1,2])
    df2  = pd.read_csv('ResultsOgden/' +InitialGuess +'/'  + str(NumberElementsTested[1]) + 'Elements_' + LoadCase, sep=',', decimal='.', usecols=[1,2])
    df3  = pd.read_csv('ResultsOgden/' +InitialGuess +'/'  + str(NumberElementsTested[2]) + 'Elements_' + LoadCase, sep=',', decimal='.', usecols=[1,2])
    df5  = pd.read_csv('ResultsOgden/' +InitialGuess +'/'  + str(NumberElementsTested[3]) + 'Elements_' + LoadCase, sep=',', decimal='.', usecols=[1,2])
    df10 = pd.read_csv('ResultsOgden/' +InitialGuess +'/'  + str(NumberElementsTested[4]) + 'Elements_' + LoadCase, sep=',', decimal='.', usecols=[1,2])
    df20 = pd.read_csv('ResultsOgden/' +InitialGuess +'/'  + str(NumberElementsTested[5]) + 'Elements_' + LoadCase, sep=',', decimal='.', usecols=[1,2])

    ax = fig.add_subplot(1, 3, LoadCases.index(LoadCase)+1)
    ax.cla()
    ax.plot(df[0], df[1],  color = 'k', linestyle = '-')
    ax.plot(df1.Stretches, df1.P,  color = 'r', linestyle = '--', marker = 'o', markersize = 8, fillstyle='none', label=' 1 Element')
    ax.plot(df2.Stretches, df2.P,  color = 'g', linestyle = '--', marker = 'o', markersize = 8, fillstyle='none', label=' 2 Elements')
    ax.plot(df3.Stretches, df3.P,  color = 'b', linestyle = '--', marker = 'o', markersize = 8, fillstyle='none', label=' 3 Elements')
    ax.plot(df5.Stretches, df5.P,  color = 'y', linestyle = '--', marker = 'o', markersize = 8, fillstyle='none', label=' 5 Elements')
    ax.plot(df10.Stretches, df10.P,  color = 'c', linestyle = '--', marker = 'o', markersize = 8, fillstyle='none', label='10 Elements')
    ax.plot(df10.Stretches, df20.P,  color = 'm', linestyle = '--', marker = 'o', markersize = 8, fillstyle='none', label='20 Elements')

    ax.set_xlabel('Stretch ratio (-)')
    ax.set_ylabel('Stress (kPa)')
    ax.legend(loc='upper left', frameon=True, framealpha=1)
    plt.title(LoadCase)

plt.savefig(os.path.join('Comparison' + '.pdf'))

plt.show()
```
