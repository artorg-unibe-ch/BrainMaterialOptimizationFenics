```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
# --------------------------------
# Path Settings
# --------------------------------

ModelsNames = ['Linear', 'St-Venant Kirchoff','Incompressible NH','Compressible NH']
Results     = 'FE_Stresses'

Xaxis = 'Stretches'
Yaxis = 'Pzz'

LinPath = os.path.join('Results', ModelsNames[0], Results)
SVKPath = os.path.join('Results', ModelsNames[1], Results)
# INHPath = os.path.join('Results', ModelsNames[2], Results)
CNHPath = os.path.join('Results', ModelsNames[3], Results)

# Import data
Lindf = pd.read_pickle(LinPath)
SVKdf = pd.read_pickle(SVKPath)
# INHdf = pd.read_pickle(INHPath)
CNHdf = pd.read_pickle(CNHPath)

# Convert to dataframe
Lindf   = pd.DataFrame(Lindf)
SVKdf   = pd.DataFrame(SVKdf)
# INHdf   = pd.DataFrame(INHdf)
CNHdf   = pd.DataFrame(CNHdf)

# Plots
plt.rc('figure', figsize=[12,7])
fig, ax = plt.subplots(1, 1)
Lindf.plot(ax = ax, x = Xaxis, y = Yaxis, color = 'red', linestyle = '-', label = ModelsNames[0])
SVKdf.plot(ax = ax, x = Xaxis, y = Yaxis, color = 'blue', linestyle = '-', label = ModelsNames[1])
# INHdf.plot(ax = ax, x = Xaxis, y = Yaxis, color = 'green', linestyle = '-', label = ModelsNames[2])
CNHdf.plot(ax = ax, x = Xaxis, y = Yaxis, color = 'black', linestyle = '-', label = ModelsNames[3])

ax.set_xlabel('Stretch ratio')
ax.set_ylabel('Nominal Stress (kPa)')
# ax.set_xlim(df_FE[Xaxis][0],df_FE[Xaxis][num_steps])
# ax.set_ylim(df_FE[Yaxis][0]*1.05,df_FE[Yaxis][num_steps]*1.05)
ax.xaxis.set_major_locator(plt.MultipleLocator(0.5))
ax.yaxis.set_major_locator(plt.MultipleLocator(100))
ax.legend(loc='upper left', frameon=True, framealpha=1)
plt.savefig(os.path.join('Results', Results + 'ModelsComparison.pdf'))
```
